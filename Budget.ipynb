```python
class Budget:
    def __init__(self, income):
        self.income = income
        self.categories = {}
        self.expenses = {}
    def add_category(self, name, limit):
        self.categories[name] = limit
        self.expenses[name] = []
    def log_expense(self, category, amount, description):
        if category not in self.categories:
            self.add_category(category, 0)
        self.expenses[category].append((amount, description))
    def remaining_budget(self, category):
        spent = sum(e[0] for e in self.expenses.get(category, []))
        return self.categories.get(category, 0) - spent
    def summary(self):
        result = {}
        for cat in self.categories:
            spent = sum(e[0] for e in self.expenses.get(cat, []))
            result[cat] = {
                "spent": spent,
                "remaining": self.remaining_budget(cat)
            }
        return result
```
