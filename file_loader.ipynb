```python
def load_expenses_from_file(filename):
    expenses = []
    with open(filename, 'r') as file:
        for line in file:
            line = line.strip()
            if not line:
                continue
            parts = line.split(',')
            if len(parts) != 3:
                print(f"Skipping invalid line: {line}")
                continue
            category, amount, description = parts
            try:
                amount = float(amount.replace('$', ''))
            except ValueError:
                print(f"Invalid amount: {amount}")
                continue
            expenses.append((category.strip(), amount, description.strip()))
    return expenses
```
