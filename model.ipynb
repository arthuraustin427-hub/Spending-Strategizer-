```python
def predict_category(text):
    text = text.lower()
    food_keywords = [
        "burger","pizza","coffee","sandwich","fries","chicken","taco",
        "burrito","pasta","salad","steak","sushi","ramen","noodles",
        "drink","soda","juice","tea","latte","espresso","bagel",
        "donut","dessert","ice cream","cake","cookie","breakfast",
        "lunch","dinner","restaurant","cafe","mcdonald","chipotle",
        "starbucks","wendy","subway"
    ]
    transport_keywords = ["uber","bus","gas","train","taxi","lyft"]
    entertainment_keywords = ["movie","game","netflix","concert","spotify"]
    if any(word in text for word in food_keywords):
        return "Food"
    elif any(word in text for word in transport_keywords):
        return "Transport"
    elif any(word in text for word in entertainment_keywords):
        return "Entertainment"
    return "Misc"
```
